```
import pandas as pd
```

```
#loading data
df=pd.read_csv('http://13.234.66.67/summer19/datasets/social.csv')
```

```
#schema 
df.info
```

```
# top 5 rows
df.head()
```

```
#removing userid
filterdata=df.iloc[0:,2:].values
```

```
#now we can apply trainin gteat
features=filterdata[0:,0:2]
```

```
#label. finding
label=filterdata[0:,2]
```

```
#traininig and testing
from sklearn.model_selection import train_test_split
```

```
X,x,Y,y=train_test_split(features,label,test_size=0.2)
```

```
#now appply featurs scalling
from sklearn.preprocessing import StandardScaler
```

```
sc=StandardScaler()
```

```
#apply scalling in training data only
X=sc.fit_transform(X)
```

```
X
```

```
#testing data scaling
x=sc.fit_transform(x)
```

```
#knn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```
#random forest calling with number of decisoin TREE
Rclf=RandomForestClassifier(n_estimators=20)
```

```
# now training
trained=Rclf.fit(X,Y)
```

```
# now predicting 
output=trained.predict(x)
```
